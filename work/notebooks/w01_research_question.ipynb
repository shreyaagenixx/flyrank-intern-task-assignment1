## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# Why these numbers justify the lane: ~X% of the inventory is actively decaying, indicating a significant maintenance burden that cannot be efficiently handled manually. The age disparity (e.g., declining pages being significantly older on average) proves that historical content decay is a real, measurable signal in this dataset.
import os
import sys
import subprocess
import pandas as pd

# 1. Handle Google Colab environment setup
IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    # Clone the repository if it hasn't been cloned yet
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
else:
    # Local setup: move up directories to find data/raw if running deep in notebooks/
    while not os.path.isdir("data/raw") and os.getcwd() != "/":
        os.chdir("..")

# Verify path matches
data_path = "data/raw/content_refresh_anonymized.csv"
assert os.path.exists(data_path), f"Starter CSV not found at {os.path.abspath(data_path)}"

# 2. Load the starter dataset
df = pd.read_csv(data_path)

# 3. Calculate your 3 key metrics
total_pages = len(df)
declining_pages = df['trend_direction'].value_counts().get('down', 0)
declining_rate = (declining_pages / total_pages) * 100

avg_age_declining = df[df['trend_direction'] == 'down']['content_age_days'].mean()
avg_age_stable = df[df['trend_direction'] != 'down']['content_age_days'].mean()

print(f"--- SUPPORTING METRICS FROM STARTER DATA ---")
print(f"1. Total Pages Analyzed: {total_pages:,}")
print(f"2. Overall Declining Rate: {declining_rate:.2f}% of pages are actively degrading.")
print(f"3. Avg Content Age (Declining Pages): {avg_age_declining:.1f} days")
print(f"4. Avg Content Age (Stable/Growing Pages): {avg_age_stable:.1f} days")
```
